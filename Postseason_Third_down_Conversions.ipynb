```python
import numpy as np
import pandas as pd 
import math
import matplotlib.pyplot as plt 
import io
```

```python
from google.colab import files

nfl = files.upload() 
```

This is pulled from
https://www.espn.com/nfl/stats/team/_/view/offense/stat/downs/table/miscellaneous/sort/fourthDownAttempts/dir/desc 

```python
# Show spreadsheet of stats for postseason teams

df_nfl = pd.read_csv(io.BytesIO(nfl['NFL_Teams_Postseason_Stats_2021.csv']))
df_nfl
```

```python
# Display Histogram of all columns

df_nfl.hist(bins=5, figsize=(15, 10), grid=False)
plt.show()
```

```python
# Show and penalties in yards as boxplot

df_nfl.boxplot(grid=False, column= "Penalties YDS")
plt.show()
```

```python
# Show and Total penalties per team as boxplot

df_nfl.boxplot(grid=False, column= "Penalties TOTAL")
plt.show()
```
